```python
import os # import os module for OS utilities
from Bio import Entrez, SeqIO
Entrez.email = "localuser@localhost.localdomain" #Mandatory! Can be replaced with your personal email.
#creating EFetch handle
filename = "Nostoc Sp. complete genome record.gbk" #up to you, but Genbank files always use *.gb or *.gbk 
#creating EFetch handle
if not os.path.isfile(filename):
    # Downloading file...
    efetch_handle = Entrez.efetch(db="nuccore", id="142013", rettype="gb", retmode="text")
    output_file_handle = open(filename, "w")
    output_file_handle.write(efetch_handle.read())
    output_file_handle.close()
    efetch_handle.close()
    print((str(filename)) + " Saved")
else:
    print((str(filename)) + " is existing. File cannot be saved!")

GoI = "cytA" #change " CytA" here with your GoI
for genome in SeqIO.parse("Nostoc Sp. complete genome record.gbk","genbank"):
    for gene in genome.features:
        if gene.type != "CDS":
            continue
        if "gene" not in gene.qualifiers:
            continue
        if GoI not in gene.qualifiers["gene"][0]: 
            continue
        gene_seq = gene.extract(genome.seq)
GoI_fasta = ">%s|%s\n%s" % (GoI, genome.description, gene_seq)
print(GoI_fasta)
```

```python
from Bio.Blast import NCBIWWW
fasta_seq = open("Nostoc Sp. cytA.fasta").read()
PROGRAM = "blastn"
DATABASE = "nt"
SEQUENCE = fasta_seq
BLAST_handle = NCBIWWW.qblast(PROGRAM, DATABASE, SEQUENCE)
with open("BLAST result.xml","w+") as BLAST_result:
    BLAST_result.write(BLAST_handle.read())
    BLAST_handle.close()
```

```python
from Bio.Blast import NCBIXML
BLAST_result_handle = open("BLAST result.xml", "r")
BLAST_record_reads = NCBIXML.read(BLAST_result_handle )
E_VALUE_THRESHOLD = 0.0001
No_of_seqs = 0
for Alignment in BLAST_record_reads.alignments:
    for HSP in Alignment.hsps: #HSP = High-scoring Segment Pair
        No_of_seqs += 1
        if HSP.expect < E_VALUE_THRESHOLD:
            print("\n")
            print("******* BLAST Alignment Report *******")
            print("Sequence:", Alignment.title)
            print("Length:", Alignment.length)
            print("E Value:", HSP.expect)
            print(HSP.query[0:100] + "...")
            print(HSP.match[0:100] + "...")
            print(HSP.sbjct[0:100] + "...")
print("\n" + "There are", No_of_seqs, "sequences in the BLAST OUTPUT")
```
